##  Step 3: Data Analysis

```python
#What is the proportion of patients with Alzheimer's vs without in this dataset?
import matplotlib.pyplot as plt

df['Diagnosis'].value_counts().plot(kind='bar', color=['steelblue', 'salmon'])
plt.title('Alzheimer Distribution')
plt.xticks([0, 1], ['No Alzheimer', 'Alzheimer'], rotation=0)
plt.ylabel('Number of Patients')
plt.show()
```

```python
df.head()
```

```python
df['Diagnosis'].value_counts()
```

```python
df.groupby('Diagnosis')['Age'].mean().T.plot(kind='bar', color=['steelblue', 'salmon'])
plt.title('Healthy vs Alzheimer')
plt.xticks(rotation=0)
plt.legend(['Healthy', 'Alzheimer'])
plt.tight_layout()
plt.show()
```

```python
categorical_cog = ['Gender', 'EducationLevel', 'Ethnicity']

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for i, col in enumerate(categorical_cog):
    crosstab = pd.crosstab(df[col], df['Diagnosis'], normalize='index')
    crosstab.plot(kind='bar', ax=axes[i], color=['steelblue', 'salmon'])
    
    axes[i].set_title(col)
    axes[i].set_xlabel('')
    axes[i].tick_params(axis='x', rotation=0)

plt.suptitle('', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
crosstab = pd.crosstab(df['Smoking'], df['Diagnosis'], normalize='index')

crosstab.plot(kind='bar')

plt.title('Smoking vs Diagnosis')
plt.xlabel('Smoking')
plt.ylabel('Proportion')
plt.xticks(rotation=0)
plt.legend(['Healthy', 'Alzheimer'])

plt.tight_layout()
plt.show()
```

```python
numerical_cog = ['BMI','AlcoholConsumption','PhysicalActivity','DietQuality','SleepQuality']
df.groupby('Diagnosis')[numerical_cog].mean().T.plot(kind='bar', color=['steelblue', 'salmon'])
plt.title('Healthy vs Alzheimer')
plt.xticks(rotation=0)
plt.legend(['Healthy', 'Alzheimer'])
plt.tight_layout()
plt.show()
```

```python
categorical_cog = [
    'FamilyHistoryAlzheimers', 'CardiovascularDisease', 'Diabetes',
    'Depression', 'HeadInjury', 'Hypertension'
]

fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

for i, col in enumerate(categorical_cog):
    crosstab = pd.crosstab(df[col], df['Diagnosis'], normalize='index')
    crosstab.plot(kind='bar', ax=axes[i], color=['steelblue', 'salmon'])
    
    axes[i].set_title(col)
    axes[i].set_xlabel('')
    axes[i].tick_params(axis='x', rotation=0)

plt.suptitle('Health Factors vs Diagnosis', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
clinical = ['SystolicBP', 'DiastolicBP', 'CholesterolTotal', 'CholesterolLDL', 'CholesterolHDL', 'CholesterolTriglycerides']

#f.groupby('Diagnosis')[clinical].mean()
df.groupby('Diagnosis')[clinical].mean().T.plot(kind='bar', color=['steelblue', 'salmon'])
plt.title('Clinical Measurements: Healthy vs Alzheimer')
plt.xticks(rotation=45)
plt.legend(['Healthy', 'Alzheimer'])
plt.tight_layout()
plt.show()
```

```python
df.groupby('Diagnosis')[clinical].mean()
```

```python
numerical_cog = ['MMSE', 'FunctionalAssessment', 'ADL']
df.groupby('Diagnosis')[numerical_cog].mean()
```

```python
df.groupby('Diagnosis')[numerical_cog].mean().T.plot(kind='bar', color=['steelblue', 'salmon'])
plt.title('Cognitive Scores: Healthy vs Alzheimer')
plt.xticks(rotation=0)
plt.legend(['Healthy', 'Alzheimer'])
plt.tight_layout()
plt.show()
```

```python
categorical_cog = ['MemoryComplaints', 'BehavioralProblems']

for col in categorical_cog:
    print(col)
    print(pd.crosstab(df[col], df['Diagnosis'], normalize='index'))
    print()
```

```python
categorical_cog = ['MemoryComplaints', 'BehavioralProblems']

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

for i, col in enumerate(categorical_cog):
    crosstab = pd.crosstab(df[col], df['Diagnosis'], normalize='index')
    crosstab.plot(kind='bar', ax=axes[i], color=['steelblue', 'salmon'])
    
    axes[i].set_title(col)
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Proportion')
    axes[i].tick_params(axis='x', rotation=0)

    # Optional: map 0/1 to No/Yes safely
    axes[i].set_xticks(range(len(crosstab.index)))
    axes[i].set_xticklabels(['No', 'Yes'])

plt.suptitle('Cognitive Symptoms: Healthy vs Alzheimer', fontsize=14)
plt.tight_layout()
plt.show()
```
